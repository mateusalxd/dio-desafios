# Funções utilitárias

````python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, classification_report
import matplotlib.pyplot as plt
````

````python
random_state_default = 42
````

````python
def separar_treino_teste(dados, coluna_resultado, usar_estratificar=False, **kwargs):
    x = dados.drop(coluna_resultado, axis=1)
    y = dados[coluna_resultado]
    estratificar = None
    if usar_estratificar:
        estratificar = y

    return train_test_split(x, y, random_state=random_state_default, stratify=estratificar, **kwargs)
````

````python
def treinar_modelo(classe_modelo, dados_treino, dados_treino_resultado, **kwargs):
    modelo = classe_modelo(random_state=random_state_default, **kwargs)
    modelo.fit(dados_treino, dados_treino_resultado)

    return modelo
````

````python
def gerar_grafico_roc(modelo, dados_teste, dados_teste_resultado):
    y_probs = modelo.predict_proba(dados_teste)[:, 1]
    fpr, tpr, _ = roc_curve(dados_teste_resultado, y_probs)

    plt.plot(fpr, tpr)
    plt.title('ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()

    print(f"AUC: {roc_auc_score(dados_teste_resultado, y_probs):.2f}")
````

````python
def gerar_grafico_precision_recall(modelo, dados_teste, dados_teste_resultado):
    y_probs = modelo.predict_proba(dados_teste)[:, 1]
    precision, recall, _ = precision_recall_curve(dados_teste_resultado, y_probs)

    plt.plot(recall, precision)
    plt.title('Precision-Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.show()
````

````python
def gerar_grafico_importancia_variaves(importancias, colunas, colunas_ignoradas=['Class']):
    colunas_consideradas = list(filter(lambda x: x not in colunas_ignoradas, colunas))

    plt.barh(colunas_consideradas, importancias)
    plt.title('Importância das Variáveis')
    plt.xlabel('Importância')
    plt.ylabel('Variável')
    plt.show()
````

# Prepação dos dados

````python
# considerar um tempo, pressupõe uma ordem, não irei considerar isso
if 'Time' in dados.columns:
    dados = dados.drop('Time', axis=1)
````

````python
import numpy as np
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

dados['Amount_log1p'] = dados['Amount'].apply(lambda x: np.log1p(x))
dados['Amount_scaler'] = scaler.fit_transform(dados[['Amount']])
````

````python
todas_colunas = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
        'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
        'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount',
        'Amount_log1p', 'Amount_scaler', 'Class']
colunas_originais  = list(filter(lambda x: x not in ['Amount_log1p', 'Amount_scaler'], todas_colunas))
````

## Vários modelos com argumentos mínimos e todas as colunas

````python
x_treino, x_teste, y_treino, y_teste = separar_treino_teste(dados[todas_colunas], 'Class', True, test_size=0.3)
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
````

````python
# https://scikit-learn.org/1.6/modules/generated/sklearn.linear_model.LogisticRegression.html
parametros = {
    'max_iter': 1000,
    # 'class_weight': 'balanced',
    'n_jobs': -1,
}
modelo = treinar_modelo(LogisticRegression, x_treino, y_treino, **parametros)
predicao = modelo.predict(x_teste)
print(classification_report(y_teste, predicao))
````

````python
# https://scikit-learn.org/1.6/modules/generated/sklearn.ensemble.HistGradientBoostingClassifier.html
parametros = {
    'max_iter': 1000,
    'scoring': 'precision',
    'class_weight': 'balanced',
}
modelo = treinar_modelo(HistGradientBoostingClassifier, x_treino, y_treino, **parametros)
predicao = modelo.predict(x_teste)
print(classification_report(y_teste, predicao))
````

````python
# treinamento lento
# https://scikit-learn.org/1.6/modules/generated/sklearn.ensemble.RandomForestClassifier.html
parametros = {
    # 'n_estimators': 50,
    'max_depth': 10,
    'class_weight': 'balanced',
    'n_jobs': -1,
}
modelo = treinar_modelo(RandomForestClassifier, x_treino, y_treino, **parametros)
predicao = modelo.predict(x_teste)
print(classification_report(y_teste, predicao))
````

````python
# https://xgboost.readthedocs.io/en/release_3.2.0/parameter.html
parametros = {
    'scale_pos_weight': 10,
    'eval_metric': 'aucpr',
}
modelo = treinar_modelo(XGBClassifier, x_treino, y_treino, **parametros)
predicao = modelo.predict(x_teste)
print(classification_report(y_teste, predicao))
````

## Modelo XGBClassifier com oversampling

````python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=random_state_default)
dados_oversampling, _ = smote.fit_resample(dados[colunas_originais], dados['Class'])

dados_oversampling['Amount_log1p'] = dados_oversampling['Amount'].apply(lambda x: np.log1p(x))
dados_oversampling['Amount_scaler'] = scaler.fit_transform(dados_oversampling[['Amount']])
````

````python
x_treino, x_teste, y_treino, y_teste = separar_treino_teste(dados_oversampling[todas_colunas], 'Class', False, test_size=0.3)

parametros = {
    # 'max_depth': 2,
    'eval_metric': 'logloss',
}
modelo = treinar_modelo(XGBClassifier, x_treino, y_treino, **parametros)
predicao = modelo.predict(x_teste)
print(classification_report(y_teste, predicao))
````

````python
gerar_grafico_roc(modelo, x_teste, y_teste)
````

````python
gerar_grafico_precision_recall(modelo, x_teste, y_teste)
````

````python
gerar_grafico_importancia_variaves(modelo.feature_importances_, todas_colunas)
````

## Modelo XGBClassifier com undersampling

````python
dados_classe_1 = dados[dados['Class'] == 1]
dados_classe_0 = dados[dados['Class'] == 0].sample(len(dados_classe_1), random_state=random_state_default)
dados_undersampling = pd.concat([dados_classe_0, dados_classe_1])
````

````python
x_treino, x_teste, y_treino, y_teste = separar_treino_teste(dados_undersampling[todas_colunas], 'Class', True, test_size=0.3)

parametros = {
    'max_depth': 0,
    'eval_metric': 'logloss',
}
modelo = treinar_modelo(XGBClassifier, x_treino, y_treino, **parametros)
predicao = modelo.predict(x_teste)
print(classification_report(y_teste, predicao))
````

````python
gerar_grafico_roc(modelo, x_teste, y_teste)
````

````python
gerar_grafico_precision_recall(modelo, x_teste, y_teste)
````

````python
gerar_grafico_importancia_variaves(modelo.feature_importances_, todas_colunas)
````

# Resultado

Os modelos **LogisticRegression**, **HistGradientBoostingClassifier** e **RandomForestClassifier**, usando parâmetros básicos, não apresentaram valores de **F1-score** acima de 79% para a classe de interesse, diferente do **XGBClassifier** que chegou em 84%.

Utilizando a abordagem de **oversampling**, o modelo apresentou **overfitting**, porém usando **undersampling**, o modelo conseguiu um **F1-score** de 94% com precision e recall de 96% e 93% respectivamente.
